```python
import csv
import datetime
import pandas as pd 

from operator import itemgetter

pd.set_option('display.max_rows', None)

tod = datetime.datetime.now()
last_three_weeks = (tod - datetime.timedelta(days = 21)).date()
last_two_weeks = (tod - datetime.timedelta(days = 14)).date()
last_week = (tod - datetime.timedelta(days = 7)).date()

by_date = []
test_three = {
        'name': 'Three Weeks',
        'last_seen': last_three_weeks - datetime.timedelta(days=1)
    }
test_two =     {
        'name': 'Two Weeks',
        'last_seen': last_two_weeks - datetime.timedelta(days=1)
    }
test_one = {
        'name': 'One Week',
        'last_seen': last_week - datetime.timedelta(days=1)
    }        
pax_last_seen = [
    test_one,
    test_two,
    test_three
]


with open('bb.csv', newline='', encoding='utf-8-sig') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        all_pax = row['PAX'].split(',')
        qic = row['QIC'].replace('and', ',').split(',')
        all_pax += qic
        when = datetime.datetime.strptime(row['When'], '%m/%d/%Y').date()
        
        for pax in all_pax:
            pax = pax.replace('(RESPECT)', '').replace(' ', '').lower()
            updated = False
            match = False
            for i, record in enumerate(pax_last_seen):
                if record['name'] == pax and not updated:
                    match = True
                    last_seen = pax_last_seen[i]['last_seen']
                    if when > last_seen:
                        pax_last_seen[i]['last_seen'] = when
                        updated = True
            if not match and not updated:
                new_record = {
                    'name': pax,
                    'last_seen': when
                }
                pax_last_seen.append(new_record)
                
```

```python
missing_one_week = [pax for pax in pax_last_seen if pax['last_seen'] < last_week and pax['last_seen'] > last_two_weeks]
pd.DataFrame(sorted(missing_one_week, key = lambda i: i['name']))
```

```python
missing_two_week = [pax for pax in pax_last_seen if pax['last_seen'] < last_two_weeks and pax['last_seen'] > last_three_weeks]
pd.DataFrame(sorted(missing_two_week, key = lambda i: i['name']))
```

```python
missing_three_weeks_or_more = [pax for pax in pax_last_seen if pax['last_seen'] < last_three_weeks]
pd.DataFrame(sorted(missing_three_weeks_or_more, key = lambda i: i['name']))
```
